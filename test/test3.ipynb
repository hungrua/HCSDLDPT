```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft
audio_file = 'Eng01.wav'
y, sr = librosa.load(audio_file)
window = np.hanning(len(y))

# p dng ca s Hanning vo tn hiu m thanh
windowed_y = y * window

# Hin th tn hiu m thanh sau khi p dng ca s Hanning
plt.figure(figsize=(10, 6))
librosa.display.waveshow(windowed_y, sr=sr, color='yellow')
plt.title('Signal after Applying Hanning Window')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
librosa.display.waveshow(y,sr=sr,color="yellow")
```

```python
ft = fft(y)
magnitude = np.absolute(ft)
frequency = np.linspace(0,sr,len(magnitude))
```

```python
plt.figure(figsize=(18,8))
plt.plot(frequency[:200],magnitude[:200])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
```
