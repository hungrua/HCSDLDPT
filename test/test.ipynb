```python
import librosa
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import IPython.display as ipd
AUDIO_FILE = '../Eng_audio/Eng21.wav'
samples,sr= librosa.load(AUDIO_FILE,duration=7)
ipd.Audio(AUDIO_FILE)
print(samples,sr)
# plt.figure(figsize=(14,5))
librosa.display.waveshow(samples,sr=sr,color="blue",alpha=0.5)
samples.shape
rms = librosa.feature.rms(y=samples)
rms.shape
```

```python
samples.shape
```

```python
frame_length = int(sr * 25/1000)
# hop_length = int(sr * 0.01) 
mfccs = librosa.feature.mfcc(y=samples,n_mfcc=12,sr=sr)
```

```python
mfccs.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(mfccs,x_axis="time",sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
delta_mfccs = librosa.feature.delta(mfccs)
delta2_mfccs = librosa.feature.delta(mfccs,order=2)
```

```python
delta_mfccs.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delta_mfccs,x_axis="time",sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(delta2_mfccs,x_axis="time",sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
comprehensive_mfccs = np.concatenate((mfccs,delta_mfccs,delta2_mfccs))
```
